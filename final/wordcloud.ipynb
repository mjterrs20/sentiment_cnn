```python
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as matplt
```

```python
# Worcdcloud keseluruhan
text = df['text']
wordcloud = WordCloud(background_color="black").generate(' '.join(text))

matplt.imshow(wordcloud, interpolation='bilinear')
matplt.axis("off")
matplt.show()
```

```python
# where label = 1
matplt.figure(figsize = (10,10))
wc = WordCloud(max_words = 2000, width = 1600, height = 800).generate(" ".join(df[df.label == 1].text))
matplt.imshow(wc, interpolation = 'bilinear')
```
